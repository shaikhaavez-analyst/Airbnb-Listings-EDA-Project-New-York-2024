**Task 4: Data Cleaning**

```python
# Finding Null Values
data.isnull().sum()

#droping null values
data.dropna(inplace=True)

# data['price'].fillna(data['price'].mean())
```

```python
# dealing with duplicates rows
data.duplicated().sum()

# data[data.duplicated()]


# deleting all duplicated rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# type casting 
# changing data types 

data.dtypes
```

```python
#changing the datatype of "id" column from float to object(string)
data['id'] = data['id'].astype(object)


#changing the datatype of "host_id" column from float to object(string)
data['host_id'] = data['host_id'].astype(object)
```

# EDA (Explorataory Data Analysis)

**Task 5: Data Analysis**

*Univariate Analysis*

```python
# idenfying outliers in price

data[data['price'] < 1500]

sns.boxplot(data=data[data['price'] < 1500], x='price')
```

```python
#Price distribuion Using Histogram

plt.figure(figsize=(8,5))
sns.histplot(data=data[data['price'] < 1500], x='price', bins=80)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
data.dtypes
```

```python
# idenfying outliers in reviews_per_month

data[data['reviews_per_month'] < 10]

sns.boxplot(data=data[data['reviews_per_month'] < 10], x='reviews_per_month')
```

```python
# reviews_per_month Distribution

plt.figure(figsize=(8,5))
sns.histplot(data=data[data['reviews_per_month'] < 10], x='reviews_per_month')
plt.title('reviews_per_month Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = data[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```
